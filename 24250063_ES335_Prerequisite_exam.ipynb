Question 1: Number of multiplication and additions for matrices of shape A(n,k) and B(k,m)

The shape of the resulting matrix C will be (n,m) by rules of matrix miltiplication. Each of element of C(C<sub>ij</sub>) is the dot product of A<sub>ij</sub> and B<sub>ij</sub>. So total number of elemnts in C is n×m so total number of multiplications is n×m×k.

For finding out additions, we need to sum the dot products find the elements C<sub>ij</sub>. So for each C<sub>ij</sub> we require to add k products so k-1 additions:
$$
C_{ij} = \sum_{l=1}^{k} A_{il} \cdot B_{lj}
$$
so total additions are : n×m×(k-1)

This is a reference code, I have to practice more with python classes and objects

```python
class Student:
    def __init__(self, roll_number, name, marks):
        self.roll_number = roll_number
        self.name = name
        self.marks = marks

    def __repr__(self):
        return f"Student(roll_number={self.roll_number}, name='{self.name}', marks={self.marks})"


class Branch:
    def __init__(self, name):
        self.name = name
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def __repr__(self):
        return f"Branch(name='{self.name}', students={self.students})"


class Year:
    def __init__(self, year):
        self.year = year
        self.branches = []

    def add_branch(self, branch):
        self.branches.append(branch)

    def __repr__(self):
        return f"Year(year={self.year}, branches={self.branches})"


class Database:
    def __init__(self):
        self.years = []

    def add_year(self, year):
        self.years.append(year)

    def __repr__(self):
        return f"Database(years={self.years})"


# Example
db = Database()


student1 = Student(1, 'A', {'Maths': 100, 'English': 70})
student2 = Student(2, 'B', {'Maths': 85, 'English': 90})


branch1 = Branch('Branch 1')
branch1.add_student(student1)
branch1.add_student(student2)


year2022 = Year(2022)
year2022.add_branch(branch1)
db.add_year(year2022)

print(db)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100, 0.5)
y1 = x
y2 = x**2
y3 = (x**3)/100
y4 = np.sin(x)
y5 = np.sin(x)/x
y6 = np.log(x)
y7 = np.exp(x)

plt.style.use('seaborn-darkgrid')

fig, axs = plt.subplots(3, 2, figsize=(15, 20))

fig.tight_layout(pad=5.0)

axs[0, 0].plot(x, y1, label='y = x', color='b')
axs[0, 0].legend()

axs[0, 1].plot(x, y2, label='y = x^2', color='g')
axs[0, 1].legend()

axs[1, 0].plot(x, y3, label='y = x^3 / 100', color='r')
axs[1, 0].legend()

axs[1, 1].plot(x, y4, label='y = sin(x)', color='c')
axs[1, 1].legend()

axs[2, 0].plot(x, y5, label='y = sin(x) / x', color='m')
axs[2, 0].legend()

axs[2, 1].plot(x, y6, label='y = log(x)', color='y')
axs[2, 1].legend()

fig_exp, ax_exp = plt.subplots(figsize=(15, 5))
ax_exp.plot(x, y7, label='y = e^x', color='k')
ax_exp.legend()

plt.show()
```

To compute the index of the minimum element in a NumPy array

```python
import numpy as np
def min(array):
    min_index = 0
    min_value = array[0]

    for i in range(1, len(array)):
        if array[i] < min_value:
            min_value = array[i]
            min_index = i

    return min_index

arr = np.random.rand(10)
print("Array:", arr)
min_index = min(arr)
print("min index:", min_index)
np_min_index = np.argmin(arr)
print("Numpy argmin index:", np_min_index)

assert min_index == np_min_index, "The min function did not match np.argmin"
print("The min function matches np.argmin")
```
